```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Pie chart
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels)
```

```python
# Increase the size of pie chart
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels, radius = 2)
plt.show()
```

```python
# To make the equal size of pie chart if it is not a perfect circular
plt.pie(df.temperature, labels = df.city, autopct = '%0.2f%%')
plt.show()
```

```python
# To show the percentage
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels = exp_labels, autopct = '%0.2f%%')
plt.axis('equal')
plt.show()
```

```python
# To show shadow
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels = exp_labels, autopct = '%0.2f%%', shadow = True)
plt.show()
```

```python
# Using explode arrays
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels = exp_labels, autopct = '%0.2f%%', explode = [0,0.1,0,0,0], radius = 1)
plt.show()
```

```python
# To align according to angle
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels = exp_labels, autopct = '%0.2f%%', startangle = 90)
plt.show()
```
