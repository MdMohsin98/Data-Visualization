```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days=[1,2,3,4,5,6,7]
max_t=[32,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
# Plotting more than one graphs 
plt.plot(days, max_t)
plt.plot(days, min_t)
plt.plot(days, avg_t, linewidth = 5, color = 'black', label = 'Avg. Temp.')
plt.xlabel('Days')
plt.ylabel('Min, Avg, Max')
plt.title("Temperatures")
```

```python
# Using Legend function : 
plt.plot(days, max_t, label = 'Max. Temp.')
plt.plot(days, min_t, label = 'Min. Temp.')
plt.plot(days, avg_t, linewidth = 5, color = 'black', label = 'Avg. Temp.')
plt.xlabel('Days')
plt.ylabel('Min, Avg, Max')
plt.title("Temperatures")
plt.legend()
```

```python
# Changing the location of legend 
plt.plot(days, max_t, label = 'Max. Temp.')
plt.plot(days, min_t, label = 'Min. Temp.')
plt.plot(days, avg_t, linewidth = 5, color = 'black', label = 'Avg. Temp.')
plt.xlabel('Days')
plt.ylabel('Min, Avg, Max')
plt.title("Temperatures")
plt.legend(loc = "upper right")
```

```python
# Changing the location of legend and replace it at its best place
plt.plot(days, max_t, label = 'Max. Temp.')
plt.plot(days, min_t, label = 'Min. Temp.')
plt.plot(days, avg_t, linewidth = 5, color = 'black', label = 'Avg. Temp.')
plt.xlabel('Days')
plt.ylabel('Min, Avg, Max')
plt.title("Temperatures")
plt.legend(loc = "best")
```

```python
# Adding shadow property in our legend
plt.plot(days, max_t, label = 'Max. Temp.')
plt.plot(days, min_t, label = 'Min. Temp.')
plt.plot(days, avg_t, linewidth = 5, color = 'black', label = 'Avg. Temp.')
plt.xlabel('Days')
plt.ylabel('Min, Avg, Max')
plt.title("Temperatures")
plt.legend(loc = "best", shadow = True, fontsize = 8)
```

```python
# Adding shadow property in our legend
plt.plot(days, max_t, label = 'Max. Temp.')
plt.plot(days, min_t, label = 'Min. Temp.')
plt.plot(days, avg_t, linewidth = 5, color = 'black', label = 'Avg. Temp.')
plt.xlabel('Days')
plt.ylabel('Min, Avg, Max')
plt.title("Temperatures")
plt.legend(loc = "lower center", shadow = True, fontsize = 'small')
```

```python
# Adding grid in our graph
plt.plot(days, max_t, label = 'Max. Temp.')
plt.plot(days, min_t, label = 'Min. Temp.')
plt.plot(days, avg_t, linewidth = 5, color = 'black', label = 'Avg. Temp.')
plt.xlabel('Days')
plt.ylabel('Min, Avg, Max')
plt.title("Temperatures")
plt.legend(loc = "best", shadow = True, fontsize = 'small')
plt.grid()
```
